# Collect the Tags for all Images from API call

```python
# libraries
import requests
import pandas as pd
import numpy as np
```

```python
## api call for json data
apiURL  = "https://api.apispreadsheets.com/data/13852/"
req = requests.get(apiURL)
```

```python
if req.status_code == 200: # SUCCESS
  jsonData = req.json()
  df = pd.read_json(jsonData)
  display(df.head())
else: # ERROR
	print("Something went wrong!!")
```
